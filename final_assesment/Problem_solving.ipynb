```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies_data = pd.read_csv("cleaned_data_all_movies.csv")
movies_data
```

```python
import seaborn as sns
```

```python
movies_data['Profit'] = movies_data['BoxOffice (Crores)'] - movies_data['Budget (Crores)']
movies_data['ROI'] = (movies_data['Profit'] / movies_data['Budget (Crores)']).round(2)

```

# P1> Calculate ROI ( Return on Investment ) for each movie

	- Which are the top 3 movies with highest ROI across all industries?
	- Which are the top 3 Hindi movies with highest ROI?
	- Which are the top 3 Telugu movies with highest ROI?
	- Plot all this information in separate graphs


```python
roi=movies_data[['MovieID', 'Title', 'ROI','Language']]
roi
```

```python
TOP_3_MOVIES = roi.sort_values(by='ROI', ascending=False).head(3)
TOP_3_MOVIES
```

```python
language = 'Telugu'
top_3_telugu = roi[roi['Language'].str.lower() == language.lower()] \
                    .sort_values(by='ROI', ascending=False) \
                    .head(3)
top_3_telugu
```

```python
language = 'Hindi'
top_3_hindi = roi[roi['Language'].str.lower()==language.lower()]\
.sort_values(by='ROI',ascending=False)\
.head(3)
top_3_hindi
```

```python
plt.figure(figsize=(8, 6))
plt.bar(TOP_3_MOVIES['Title'], TOP_3_MOVIES['ROI'], color='LIGHTBLUE')
plt.title('OVERALL TOP 3 MOVIES WITH HIGHEST ROI')
plt.xlabel('TITLE')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(top_3_telugu['Title'], top_3_telugu['ROI'], color='LIGHTBLUE')
plt.title('TOP 3 RETURN OF INVESTENT : TELUGU MOVIES ')
plt.xlabel('TITLE')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(top_3_hindi['Title'], top_3_hindi['ROI'], color='LIGHTBLUE')
plt.title('TOP 3 RETURN OF INVESTENT : HINDI MOVIES ')
plt.xlabel('TITLE')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Analysis 1: Director Performance Based on Average Box Office Collection

## Description:
-This analysis ranks directors based on the average box office collection of the movies they directed. It helps identify which directors consistently deliver commercially successful films.

```python
# Group by director and calculate average box office collection
avg_boxoffice_by_director = movies_data.groupby('Director')['BoxOffice (Crores)'].mean().sort_values(ascending=False)

# Plot top 10 directors
top_directors = avg_boxoffice_by_director.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title("Top 10 Directors by Avg. Box Office Collection")
plt.xlabel("Average Box Office (Crores)")
plt.ylabel("Director")
plt.show()
```

### Analysis 2: ROI (Return on Investment) by Genre

### Description:
This analysis evaluates which genres provide the highest return relative to their production budget. ROI is calculated using the formula:

```python
# Add ROI column: (BoxOffice - Budget) / Budget
movies_data['ROI'] = (movies_data['BoxOffice (Crores)'] - movies_data['Budget (Crores)']) / movies_data['Budget (Crores)']

# Group by genre and compute average ROI
avg_roi_by_genre = movies_data.groupby('Genre')['ROI'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=avg_roi_by_genre.values, y=avg_roi_by_genre.index)
plt.title("Average ROI by Genre")
plt.xlabel("Average ROI")
plt.ylabel("Genre")
plt.show()
```
